```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Customer Churn.csv")
df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
x=df.drop('Churn_Yes',axis=1)
y=df['Churn_Yes']
print("Length of X:", len(x))
print("Length of y:", len(y))
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rf=RandomForestClassifier(random_state=42)
rf.fit(x_train,y_train)
```

```python
importances = pd.DataFrame({'Feature': x_train.columns,'Importance': rf.feature_importances_}).sort_values(by='Importance', ascending=False)
print("\nTop 10 Important Features:")
print(importances.head(10))
```

```python
selected_features = importances[importances['Importance'] >= 0.02]['Feature'].tolist()
print("Selected Features:", selected_features)
```

```python
x_selected = df[selected_features]
y = df['Churn_Yes']
x_train, x_test, y_train, y_test = train_test_split(x_selected, y, test_size=0.2, random_state=42)
```

```python
rf_selected=RandomForestClassifier(random_state=42)
rf_selected.fit(x_train,y_train)
```

```python
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(x_train, y_train)
```

```python
prediction=log_reg.predict([[24, 55.0, 0, 1,1]])
if prediction[0] == 1:
    print("Customer will CHURN.")
else:
    print("Customer will NOT churn.")
```
